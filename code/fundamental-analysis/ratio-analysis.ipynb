```python
import pandas as pd
import numpy as np

# load data
filename = 'jp_funda-input.csv' # use your path
df = pd.read_csv(filename, header=0, index_col='Ticker')
print(len(df))
```

```python
sorted_cols = sorted(list(df.columns))
for col in sorted_cols:
    print(col)
```

```python
ratios = pd.DataFrame(index=df.index)
#ratios['Ticker'] = df['Ticker'] # create new df to store ratios

print(len(ratios))
ratios.head()
```

### Asset Utilization or Turnover ratios

```python
# Total Asset Turnover
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['Total Revenue', 'Total Assets']]

df_tmp = df_tmp.fillna(0)

TAT = df_tmp['Total Revenue'].astype(int) / df_tmp['Total Assets'].astype(int)# Total Revenue / Total Assets
df_tmp["Total asset turnover"] = TAT    
df_tmp = df_tmp[['Total asset turnover']]

results = pd.concat([ratios, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

### Financial Leverage ratios

```python
# Total Debt Ratio
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['Total Liabilities Net Minority Interest', 'Total Assets']]

df_tmp = df_tmp.fillna(0)
#print(df_tmp)

TDR = df['Total Liabilities Net Minority Interest'].astype(int) / df['Total Assets'].astype(int) # Total Liabilities / Total Assets
df_tmp["Total debt ratio"] = TDR    
df_tmp = df_tmp[['Total debt ratio']]

results = pd.concat([results, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

```python
# Debt/Equity
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['Total Liabilities Net Minority Interest', "Total stockholders' equity"]]
df_tmp = df_tmp.fillna(0)
#df_tmp = df_tmp[(df_tmp["Total stockholders' equity"] != 0)]

DE = df_tmp['Total Liabilities Net Minority Interest'].astype(int) / df_tmp["Total stockholders' equity"].astype(int) # Total Liabilities / Total stockholders' equity
df_tmp["Debt equity ratio"] = DE  
df_tmp = df_tmp[['Debt equity ratio']]

results = pd.concat([results, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

```python
# Equity Ratio
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['Total Assets', "Total stockholders' equity"]]
df_tmp = df_tmp.fillna(0)

ER = df_tmp["Total stockholders\' equity"].astype(int) / df_tmp['Total Assets'].astype(int) # Total stockholders' equity / Total Assets
df_tmp["Equity ratio"] = ER  
df_tmp = df_tmp[['Equity ratio']]

results = pd.concat([results, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

```python
# Long-term Debt Ratio
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['Total Assets', "Long Term Debt"]]
df_tmp = df_tmp.fillna(0)
#df_tmp = df_tmp[df_tmp['Long Term Debt'] != '-']
df_tmp['Long Term Debt'] = np.where(df_tmp['Long Term Debt'] == '-', 0, df_tmp['Long Term Debt'])

LTDR = df_tmp['Long Term Debt'].astype(float) / df_tmp['Total Assets'].astype(float) # Long Term Debt / Total Assets
df_tmp["Long-term debt ratio"] = LTDR  
df_tmp = df_tmp[['Long-term debt ratio']]

results = pd.concat([results, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

```python
# Times Interest Earned Ratio
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['EBIT', "Interest Expense"]]

#df_tmp = df_tmp.dropna()
#df_tmp = df_tmp[df_tmp['Interest Expense'] != '-']
df_tmp = df_tmp.fillna(0)
df_tmp['Interest Expense'] = np.where(df_tmp['Interest Expense'] == '-', 0, df_tmp['Interest Expense'])

TIER = df_tmp['EBIT'].astype(float) / df_tmp['Interest Expense'].astype(float) # Earnings Before Interest and Taxes / Interest
df_tmp["Times interest earned ratio"] = TIER  
df_tmp = df_tmp[['Times interest earned ratio']]

results = pd.concat([results, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

### Profitability ratios

```python
# Gross Profit Margin
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['Gross Profit', "Total Revenue"]]
df_tmp = df_tmp.fillna(0)

GPM = df_tmp['Gross Profit'].astype(int) / df_tmp['Total Revenue'].astype(int) # Gross Profit / Total Revenue
df_tmp["Gross profit margin"] = GPM 
df_tmp = df_tmp[['Gross profit margin']]

results = pd.concat([results, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

```python
# Net Profit Margin
df_tmp = pd.DataFrame(index=df.index)
df_tmp = df[['Net Income Common Stockholders', "Total Revenue"]]
df_tmp = df_tmp.fillna(0)

NPM = df_tmp['Net Income Common Stockholders'].astype(int) / df_tmp['Total Revenue'].astype(int) # Net Income / Total Revenue
df_tmp["Net profit margin"] = NPM
df_tmp = df_tmp[['Net profit margin']]

results = pd.concat([results, df_tmp], axis=1, join='inner')

#ratios = ratios.merge(df_tmp, how='inner', left_on=['Ticker'], right_on=['Ticker']) # merge dfs
#df_out = pd.concat([ratios, df_tmp], axis=1)
#ratios[ratios.duplicated('Ticker', keep=False)]

print(results)
```

### Screening masks

```python
# Profitability
mask1 = (ratios['EPS'] > ratios['EPS'].mean())
mask2 = (ratios['ROE'] > ratios['ROE'].mean())
mask3 = (ratios['ROA'] > ratios['ROA'].mean())

# Asset utilisation
mask4 = (ratios['Total asset turnover'] > ratios['Total asset turnover'].mean())

# Financial leverage ratio
mask5 = (ratios['Total debt ratio'] < ratios['Total debt ratio'].mean())
mask6 = (ratios['Debt equity ratio'] < ratios['Debt equity ratio'].mean())
mask7 = (ratios['Equity ratio'] < ratios['Equity ratio'].mean())
mask8 = (ratios['Long-term debt ratio'] < ratios['Long-term debt ratio'].mean())
```

```python
ratios[(mask2) & (mask3) & (mask4) & (mask5) & (mask8)]
```

### Scoring

```python
# Profitability
score = pd.DataFrame(index=ratios.index)

mask1 = (ratios['EPS'] > ratios['EPS'].mean())
mask2 = (ratios['ROE'] > ratios['ROE'].mean())
mask3 = (ratios['ROA'] > ratios['ROA'].mean())

score['profit_score1'] = np.where(mask1, 1.0, 0.0)
score['profit_score2'] = np.where(mask2, 1.0, 0.0)
score['profit_score3'] = np.where(mask3, 1.0, 0.0)

# Asset utilisation
mask4 = (ratios['Total asset turnover'] > ratios['Total asset turnover'].mean())

score['asset_score1'] = np.where(mask5, 1.0, 0.0)

# Financial leverage ratio
mask5 = (ratios['Total debt ratio'] < ratios['Total debt ratio'].mean())
mask6 = (ratios['Debt equity ratio'] < ratios['Debt equity ratio'].mean())
mask7 = (ratios['Equity ratio'] < ratios['Equity ratio'].mean())
mask8 = (ratios['Long-term debt ratio'] < ratios['Long-term debt ratio'].mean())

score['leverage_score1'] = np.where(mask5, 1.0, 0.0)
score['leverage_score2'] = np.where(mask6, 1.0, 0.0)
score['leverage_score3'] = np.where(mask7, 1.0, 0.0)
```

```python
score.head()
```

```python
score['Total'] = 1/3 * score['profit_score1'] + 1/3 * score['profit_score2'] + 1/3 * score['profit_score3'] + score['asset_score1'] + 1/3 * score['leverage_score1'] + 1/3 * score['leverage_score2'] + 1/3 * score['leverage_score3']
```

```python
score['Final'] = score['Total'] / 3
```
